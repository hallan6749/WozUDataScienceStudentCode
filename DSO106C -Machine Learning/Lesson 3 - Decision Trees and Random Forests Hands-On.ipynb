Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pd.options.display.max_columns = 1000
from sklearn.model_selection import RandomizedSearchCV
```

Load in Data

```python
Titanic = sns.load_dataset('titanic')
Titanic.head()
```

```python
# embarked & embark_town are the same
del Titanic['embark_town']
Titanic
```

Dropping redundant columns

```python
# adult_male is included in the sex and who colummns
del Titanic['adult_male']
Titanic
```

```python
# survived and alive are the same column
del Titanic['alive']
Titanic
```

```python
# pclass  and class are the same column
del Titanic['class']
Titanic
```

```python
# Drop and Missing and Infinite Values
Titanic.dropna(inplace=True)
```

Recoding

```python
def sex (series):
    if series == "male":
        return 0
    if series == "female":
        return 1
Titanic['sexR'] = Titanic['sex'].apply(sex)
```

```python
#Drop sex has been recoded
del Titanic['sex']
Titanic
```

```python
Titanic.embarked.value_counts()
```

```python
def embarked (series):
    if series == "S":
        return 0
    if series == "C":
        return 1
    if series == "Q":
        return 2
Titanic['embarkedR'] = Titanic['embarked'].apply(embarked)
```

```python
#Drop embarked has been recoded
del Titanic['embarked']
Titanic
```

```python
Titanic.who.value_counts()
```

```python
def who (series):
    if series == "man":
        return 0
    if series == "woman":
        return 1
    if series == "child":
        return 2
Titanic['whoR'] = Titanic['who'].apply(who)
```

```python
del Titanic['who']
Titanic
```

```python
Titanic.info()
```

```python
Titanic.deck.value_counts()
```

```python
def deck (series):
    if series == "A":
        return 0
    if series == "B":
        return 1
    if series == "C":
        return 2
    if series == "D":
        return 3
    if series == "E":
        return 4
    if series == "F":
        return 5
    if series == "G":
        return 6
Titanic['deckR'] = Titanic['deck'].apply(deck)
```

```python
del Titanic['deck']
Titanic
```

```python
# Data is in sibsp and parch columns
del Titanic['alone']
Titanic
```

```python
Titanic.info()
```

Define x and y variables

```python
x = Titanic[['pclass', 'age', 'sibsp', 'parch', 'fare', 'sexR', 'embarkedR', 'whoR', 'deckR']]
y = Titanic ['survived']
```

Train Test split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

Decision Tree

```python
decisionTree = DecisionTreeClassifier()
decisionTree.fit(x_train, y_train)
```

```python
treePredictions = decisionTree.predict(x_test)
print(confusion_matrix(y_test, treePredictions))

# 16, 37 are correct and made a few errors
```

```python
print(classification_report(y_test, treePredictions))

# 73% accurancy, we are better at predicting survivers than non survivors
```

Ramdon Forest

```python
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
```

```python
forest = forest.predict(x_test)
print(confusion_matrix(y_test, forest))

# 15, 39 are correct and made a few errors
```

```python
print(classification_report(y_test, forest.predict(x_test)))

# 74% accurancy, we are better at predicting survivers than non survivors, 
# little bit more accurate than the Decision Tree Model
```

Overall, the decision tree and the random forest model are super similar in predictive value change for both the confusion matrix and the classification report
